# CEPH results

Here we will show the results obtained with CEPH while changin the amount of RAM, the number of core and the speed of HDDs on the servers. 


```python
df = pd.DataFrame()
for core in [8,16,24,32]:
    for ram in [32,64,128,192]:
        for mode in ["read","write","randwrite","randread"]:
            for replica in [1,2,3]:
                for run_id in [1,2,3,4,5,7]:
                    if ram==128 and core==8 and run_id>3:
                        continue
                    with open(f'8-nodes/{mode}_replica_{replica}_core_{core}_memory_{ram}_run_{run_id}.json', 'r') as file:
                        try:
                            job = json.load(file)
                        except:
                            print(f'Error opening: results/{mode}_replica_{replica}_core_{core}_memory_{ram}_run_{run_id}.json')
                            continue
                        job=job["client_stats"]
                        if "write" in mode:
                            bw=job[7]["write"]["bw_bytes"]/1000/1000/1000
                        else:
                            bw=job[7]["read"]["bw_bytes"]/1000/1000/1000
                        bw_aggregated=0
                        for i in range(0,7):
                            if "write" in mode:
                                bw_aggregated+=job[i]["write"]["bw_bytes"]/1000/1000/1000
                            else:
                                bw_aggregated+=job[i]["read"]["bw_bytes"]/1000/1000/1000
                        d={"core":core,"ram":ram,"mode":mode,"replica":replica,"run_id":run_id, "bw":bw,"bw_aggregated":bw_aggregated}
                        experiment=pd.DataFrame(d, index=[0])
                        #print(experiment['jobname'])
                        df= pd.concat([df, experiment], ignore_index=True)
```

This experiment keep the amount of RAM and core to the max while reducing the HDD bandwidth via Cgroup.

```python
df = pd.DataFrame()
for core in [32]:
    for ram in [192]:
        for mode in ["read","write","randwrite","randread"]:
            for replica in [1,2,3]:
                for run_id in [1,2,3,4,5,6,7]:
                    for speed in [100,125,150,175,200,225,250,275]:
                        with open(f'results_blkio/{mode}_replica_{replica}_core_{core}_memory_{ram}_speed_{speed}_run_{run_id}.json', 'r') as file:
                            try:
                                job = json.load(file)
                            except:
                                print(f'Error opening: results/{mode}_replica_{replica}_core_{core}_memory_{ram}_run_{run_id}.json')
                                continue
                            job=job["client_stats"]
                            #df= pd.concat([df, df_tmp], ignore_index=True)
                            if "write" in mode:
                                bw=job[5]["write"]["bw_bytes"]/1000/1000/1000
                            else:
                                bw=job[5]["read"]["bw_bytes"]/1000/1000/1000
                            bw_aggregated=0
                            for i in range(0,5):
                                if "write" in mode:
                                    bw_aggregated+=job[i]["write"]["bw_bytes"]/1000/1000/1000
                                else:
                                    bw_aggregated+=job[i]["read"]["bw_bytes"]/1000/1000/1000
                                
                            d={"core":core,"ram":ram,"mode":mode,"replica":replica,"run_id":run_id, "bw":bw,"bw_aggregated":bw_aggregated,"device_speed":speed}
                            experiment=pd.DataFrame(d, index=[0])
                            #print(experiment['jobname'])
                            df= pd.concat([df, experiment], ignore_index=True)
```

# Change the order of the experiments

Given the noise present on the measurements we tried to investigated further, the first attemp has been to change the order of the runs. First we run all the experiments in one pools, then we performed all in the experiments in round robin over the pools. This seems to not affect the results.

```python
df = pd.DataFrame()
for core in [24,32]:
    for ram in [192]:
        for mode in ["read","write"]:
            for replica in [1,2,3]:
                for run_id in [1,2,3,4,5,6,7]:
                    with open(f'results_inverted/{mode}_replica_{replica}_core_{core}_memory_{ram}_run_{run_id}.json', 'r') as file:
                        try:
                            job = json.load(file)
                        except:
                            print(f'Error opening: results/{mode}_replica_{replica}_core_{core}_memory_{ram}_run_{run_id}.json')
                            continue
                        job=job["client_stats"]
                        #df= pd.concat([df, df_tmp], ignore_index=True)
                        if "write" in mode:
                            bw=job[3]["write"]["bw_bytes"]/1024/1024
                        else:
                            bw=job[3]["read"]["bw_bytes"]/1024/1024
                        bw_aggregated=0
                        for i in range(0,3):
                            if "write" in mode:
                                bw_aggregated+=job[i]["write"]["bw_bytes"]/1024/1024
                            else:
                                bw_aggregated+=job[i]["read"]["bw_bytes"]/1024/1024
                            
                        d={"core":core,"ram":ram,"mode":mode,"replica":replica,"run_id":run_id, "bw":bw,"bw_aggregated":bw_aggregated,"stable":1}
                        experiment=pd.DataFrame(d, index=[0])
                        #print(experiment['jobname'])
                        df= pd.concat([df, experiment], ignore_index=True)
```

```python
plt.figure(figsize=(15,30),dpi=100)
for i,mode in enumerate(["read","write"]):
    for replica in [1,2,3]:
        plt.subplot(3, 2,(replica-1)*2+i+1)
        sns.violinplot(data=df[(df["mode"]==mode) & (df["replica"]==replica)], x="stable", y="bw_aggregated",hue="core",palette="tab10")
        #plt.title(f"Test type: {mode}, blocksize: {bs}")
        plt.xlabel("Normal[sx] and inverted[dx]")
        plt.ylabel("Bandwidth [MiB/s]")
        plt.ylim(1500,8000)
        plt.title(f"{mode},replica={replica}")
```

# Dept size and block size
We report the results obtained running the same tests, changing the number of client and the blocksize. 

```python
def open_result_cephfs():
    #init an empty dataframe
    df = pd.DataFrame()
    for mode in ["sequential","random"]:
            for run_id in range(1,4):
                with open(f'block_size/cephfs-libaio_{run_id}-{mode}.json', 'r') as file:
                    job = json.load(file)
                    job['run_id']=run_id
                    job['fstype']='cephfs'
                    df_tmp=json_to_dataframe(job)
                    df= pd.concat([df, df_tmp], ignore_index=True)
    #df = df.astype({'numjobs':'int'})
    return df
    
```

```python
df=open_result_cephfs()
```

```python
def plot_results_ceph(results):
    ax  = sns.relplot(
        data=results, x="numjobs", y="bandwidth",col="rw",
        hue="bs", kind="line"#, style="event", kind="line",
    )
    #ax.set(xlabel='Number of process', ylabel='BW [MiB/s]')
    ax.legend.set_title("Block size [Bytes]")
    #for a in ax.axes.flat:
        #a.set_xticks(ticks=filtered["numjobs"].unique()) # set new labels
        #a.set_xticklabels(fontsize=8, rotation=0, labels=filtered["numjobs"].unique())
        #min_y,max_y=a.get_yticks()[[0,-1]]
        #min_y=0
        #if max_y-min_y < 1000:
        #    new_ticks=list(range(int(min_y),int(max_y)+25,25))
        #    a.set_yticks(ticks=new_ticks)
        #    a.set_yticklabels(fontsize=8, rotation=0, labels=new_ticks)
        #ax.set(ylim=(0, 300))
    plt.xlabel("Number of jobs")
```

# iodepth
We report the results obtained with 2 different iodepth

```python
def open_result_iodepth():
    #init an empty dataframe
    df = pd.DataFrame()
    for run_id in range(1,4):
        with open(f'iodepth/cephfs-iodepth_{run_id}.json', 'r') as file:
            job = json.load(file)
            job['run_id']=run_id
            job['fstype']='cephfs'
            df_tmp=json_to_dataframe(job)
            df= pd.concat([df, df_tmp], ignore_index=True)
#df = df.astype({'numjobs':'int'})
    return df
df=open_result_iodepth() 
```

```python
plt.figure(figsize=(10,10))
for i,mode in enumerate(["read","write"]):
    plt.subplot(1, 2,i+1)
    sns.boxplot(data=df, x="bs", y="bandwidth",hue="iodepth")
    plt.title(f"Test type: {mode}")
    plt.xlabel("Blocksize")
    plt.ylabel("Bandwidth [MiB/s]")
```
